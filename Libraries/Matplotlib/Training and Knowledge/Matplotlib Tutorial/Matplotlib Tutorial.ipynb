### Load Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Real World Examples 

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight': 'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, label='United States', marker='.')
plt.plot(gas['Year'], gas['Canada'], label='Canada', marker='.')
plt.plot(gas['Year'], gas['South Korea'], label='South Korea', marker='.')
plt.plot(gas['Year'], gas['Australia'], label='Australia', marker='.')

## Another way to plot these countries' gas prices
# countries_to_look_at = ['USA', 'Canada', 'South Korea', 'Australia']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], label=country)

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()

gas
```

### Load FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')

bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins)

plt.ylabel('Number of Players')
plt.xlabel('Player Skills')

plt.title('Distribution of Player Skills in FIFA 2018', fontdict={'fontweight': 'bold'})

plt.show()

fifa
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aa5445']

plt.pie([left, right], labels=labels, colors=colors, autopct="%.2f %%")

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >=125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >=150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >=175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.2,0,0,.4)
plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['Fc Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels,
                    patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)
    
    # Change fill color
    box.set(facecolor='#e0e0e0')
    
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
