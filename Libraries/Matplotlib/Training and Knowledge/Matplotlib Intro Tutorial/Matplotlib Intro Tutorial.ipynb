### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1,2,3,4]
y = [2,4,6,8]

# Resize Graph--> for a better view, assign a high value to dpi. 
# dpi specifies pixel per inch
plt.figure(figsize=(5,3),dpi=300)

## Line 1

# shorthand notation --> fmt = '[color],[marker],[line]'
# plt.plot(x,y,'b^--',label='2x')

# keyword argument notation
plt.plot(x,y, label='2x', color='#Ff0000', linewidth=3, marker='.', linestyle='--', markersize=14, markeredgecolor='blue')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,label='x^2',color='green')

# plot the remainder of the graph as dashed line
plt.plot(x2[5:],x2[5:]**2,color='green',linestyle='--')

# add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis tickmarks (scale of the graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

# add a legend
plt.legend()

# save figure (you can save it on high dpi for better resolution)
plt.savefig('mygraph.png', dpi=500)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```
